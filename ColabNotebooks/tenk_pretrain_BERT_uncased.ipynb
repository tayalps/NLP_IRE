```
from tokenizers.implementations import ByteLevelBPETokenizer
from tokenizers.processors import BertProcessing


tokenizer = ByteLevelBPETokenizer(
    "./tenkBERTv2/vocab.json",
    "./tenkBERTv2/merges.txt",
)
```

```
tokenizer._tokenizer.post_processor = BertProcessing(
    ("[SEP]", tokenizer.token_to_id("[SEP]")),
    ("[CLS]", tokenizer.token_to_id("[CLS]")),
)
tokenizer.enable_truncation(max_length=512)
```

```
from transformers import BertConfig

config = BertConfig(
    vocab_size=32_000,
    max_position_embeddings=512,
    num_attention_heads=12,
    num_hidden_layers=6,
    type_vocab_size=1,
    max_seq_length=200,
)
```

```
from transformers import BertForMaskedLM

model = BertForMaskedLM(config=config)
```

```
model.num_parameters()
```

```
!pwd
```

```
from transformers import BertTokenizerFast

tokenizer = BertTokenizerFast.from_pretrained("bert-base-uncased", truncation = True, padding=True, max_length=200)
```

```
cd "/content/drive/My Drive/praveen_iiit_hyd_ire_class/ten_k_project_work"
```

```
%%time
from transformers import LineByLineTextDataset

dataset = LineByLineTextDataset(
    tokenizer=tokenizer,
    file_path="./tenk_join_8k.txt",
    block_size=100,
)
```

```
!pwd
```

```
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=True, mlm_probability=0.15
)
```

```
!pwd
```

```
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./tenkBERTv2",
    overwrite_output_dir=True,
    num_train_epochs=1,
    per_device_train_batch_size=128,
    save_steps=10_000,
    save_total_limit=2,
)

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=dataset,
    prediction_loss_only=True,
)
```

```
%%time
trainer.train()
```

```
trainer.save_model("./tenkBERTv2")
```

```
from transformers import pipeline

fill_mask = pipeline(
    "fill-mask",
    model="./tenkBERTv2",
    tokenizer = tokenizer
)
```
